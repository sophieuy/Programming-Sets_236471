```python
#2. What are the three most popular categories, by total sales?
```

```python
sales_data = pd.read_csv('fct_invoice.csv')
age_data = pd.read_json('dim_customer.json')

merged_data = pd.merge(sales_data, age_data, left_on='customer_id', right_on='id')

merged_data['total_sales'] = merged_data['quantity'] * merged_data['price']
over_45_data = merged_data[merged_data['age'] > 45]

total_sales_over_45 = over_45_data['total_sales'].sum()

print(f"Total sales attributed to customers over the age of 45: {total_sales_over_45}")
```

```python
#3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv('fct_invoice.csv')
df['total'] = df['quantity'] * df['price']

df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)
df.set_index('invoice_date', inplace=True)
daily_sales = df['total'].resample('D').sum()

plt.figure(figsize=(16, 10))
plt.plot(daily_sales.index, daily_sales.values, marker="o", linestyle='-', color='b')
plt.title('Daily Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

print("The graph below shows the daily sales across the invoice dates.")
plt.show()
```

```python
df = pd.read_csv('fct_invoice.csv')
df['total'] = df['quantity'] * df['price']

df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

df.set_index('invoice_date', inplace=True)
monthly_sales = df['total'].resample('M').sum()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, linestyle='-', color='b')
plt.title('Monthly Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)

print("The graph below shows the total monthly sales across the invoice dates.")
plt.show()
```

```python
'''1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: 
category, decade age range (e.g., 10-19, 20-29, and so forth).'''
```

```python
customers_df = pd.read_json('dim_customer.json')
sales_df = pd.read_csv('fct_invoice.csv')

sales_df['sales'] = sales_df['quantity'] * sales_df['price']
customers_df['Decade'] = (customers_df['age'] // 10) * 10

merged_df = sales_df.merge(customers_df[['id', 'Decade']], left_on='customer_id', right_on='id')

pivot_table = pd.pivot_table(merged_df, values='sales', index='category', columns='Decade', aggfunc='sum', fill_value=0)

pivot_table.columns = [f"{int(col)}-{int(col)+9}" if col != 0 else "Unknown" for col in pivot_table.columns]

display(pivot_table)
```
